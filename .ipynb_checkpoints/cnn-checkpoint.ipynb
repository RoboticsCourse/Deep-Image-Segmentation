```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import glob
import numpy as np
```

```python
inputs = [file for file in glob.glob("CSC498/input/*")]
masks = [file for file in glob.glob("CSC498/mask/*")]

input_by_id = {}
for file in inputs:
    imgid = int(file.split(" ")[-1].split(".")[0])
    input_by_id[imgid] = file
mask_by_id = {}
for file in masks:
    imgid = int(file.split("_")[-1].split("-")[0])
    mask_by_id[imgid] = file
    
test_id = [76, 17, 42]
```

```python
import cv2
```

```python
imgs = []
masks = []

for i, file in input_by_id.items():
    if i not in test_id:
        img = cv2.imread(file)
        resized = cv2.resize(img, dsize=(150, 200), interpolation=cv2.INTER_CUBIC)
        mask = cv2.imread(mask_by_id[i])[:,:,2]
        mask = cv2.resize(mask, dsize=(150, 200), interpolation=cv2.INTER_CUBIC)
        imgs.append(resized)
        masks.append(mask)
        
imgs = np.stack(imgs) / 255
masks = np.stack(masks) / 255
```

```python
imgs.shape # [29, 400, 300, 3]
```

```python
x = torch.tensor(imgs, dtype=torch.float32).transpose(1,3)
```

```python
x.shape
```

```python
masks = torch.tensor(masks, dtype=torch.float32)
```

```python
import torchvision.models
alexnet = torchvision.models.alexnet(pretrained=True)
```

```python
alexnet.features(x).shape
```

```python
8*3*5*15*200
```

```python
class Steven(nn.Module):
    def __init__(self):
        super(Steven, self).__init__()
        self.conv = nn.Conv2d(256, 8, 1)
        self.linear = nn.Linear(8*3*5, 150*200)
        
    def forward(self, x):
        alexnetout = alexnet.features(x)
        convout = self.conv(alexnetout)
        convout = convout.view(-1, 8*3*5)
        return self.linear(convout).view(-1, 200, 150)

model = Steven()
```

```python
criterion = nn.BCEWithLogitsLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
```

```python
for it in range(50):
    out = model(x)
    loss = criterion(masks.view(-1), out.view(-1))
    loss.backward()
    optimizer.step()
    optimizer.zero_grad()
    print(it, float(loss))
```
